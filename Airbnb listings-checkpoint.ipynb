```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np 
```

```python
# Importing file and casting data columns as a datetime format
listings=pd.read_csv("Listings.csv", parse_dates= ['host_since'],encoding="ISO-8859-1",low_memory=False)
listings.head()
```

```python
listings.info()
```

```python
# Filtering rows where city is Paris and keeping five columns
paris_list = (listings.query("city == 'Paris'").loc[:,["host_since","neighbourhood","city","accommodates","price"]])
paris_list.info()
```

```python
# Checking for missing values on paris_list data
paris_list.isna().sum()
```

```python
# Calculating stats for Paris listing data
paris_list.describe()
```

```python
## This is a small fraction of our data, so there is no need to drop the rows 
paris_list.query("accommodates == 0 and price == 0").count()
```

```python
# Creating a table named paris_listings_neighboourhood and calculating the mean price for each
paris_listings_neighbourhood = paris_list.groupby("neighbourhood").agg({"price":"mean"}).sort_values("price",ascending=False)
print (paris_listings_neighbourhood)
```

```python
# Creating table paris_listings_accommodations and filtering down to the most expensive neighbourhood, grouped by accommodates and corresponding mean price
paris_listings_accommodations = paris_list.query("neighbourhood =='Elysee'").groupby("accommodates").agg({"price":"mean"}).sort_values("price")
print (paris_listings_accommodations)
```

```python
# Creating a table grouped by host since to count number of new hosts with the average price for each one
paris_listings_overtime = paris_list.set_index("host_since").resample("YE").agg({"neighbourhood":"count","price":"mean"})
paris_listings_overtime.head()
```

```python
# Horizontal bar chart to visualize average price for each neighbourhood in descending order
avg_price_by_neighbourhood = paris_list.groupby("neighbourhood").agg({"price":"mean"})
avg_price_by_neighbourhood = avg_price_by_neighbourhood.sort_values(by = "price")
avg_price_by_neighbourhood.plot (kind = "barh", legend = False)
plt.title("Average Price by Neighbourhood in Paris")
plt.xlabel("Average Price")
plt.ylabel("Neighbourhood")
sns.despine()
```

```python
# Horizontal bar chart to visualize avergage price by accommodates in most expensive neighbourhood
avg_price_by_accommodates = paris_list.query("neighbourhood == 'Elysee'"). groupby("accommodates").agg({"price":"mean"})
avg_price_by_accommodates = avg_price_by_accommodates.sort_values(by = "price")
avg_price_by_accommodates.plot (kind = "barh", legend = False)
plt.title("Average Listing Price by Accommodation")
plt.xlabel("Price per Night (Euros)")
plt.ylabel("Accommodation Capacity")
sns.despine()
```

```python
# Creating line chart: new hosts over time. New hosts spiked due to demand up to 2015 and the drastically declined after due to city regulations.
paris_listings_overtime["neighbourhood"].plot(
    ylabel= "New Hosts",
    title = "New Airbnb Hosts in Paris Over time",
    c = "green"
)
sns.despine()
```

```python
# Line Chart: Average Price over time, which steadily spiked until 2009. Notice price dipped through 2014 and picked right up
# after city regulations made an impact on the number of hosts to decrease. 
paris_listings_overtime["price"].plot(
    ylabel= "Average Price (Euros)",
    title = "Average Airbnb Price in Paris Over Time"
)
sns.despine()
```

```python
# By combining both line charts, we see the impact city regulations made on Airbnb hosts and avergage prices over time.
fig, ax = plt.subplots()
ax.plot(
    paris_listings_overtime.index,
    paris_listings_overtime["neighbourhood"],
    label = "New Hosts",
    c = "green"
)
ax.set_ylabel("New Hosts")
ax2 = ax.twinx()
ax2.plot(
    paris_listings_overtime.index,
    paris_listings_overtime["price"],
    label = "Average Price",
)
ax2.set_ylim(0)
ax2.set_ylabel("Average Price")
ax.set_title("2015 Regulations Lead to Fewer New Hosts, Higher Prices")
plt.show()
```
